# Part 1: Data Analysis using NumPy
## 1. Why NumPy?
NumPy is written in C and stores data in contiguous memory blocks. This makes operations up to 50x faster than standard Python lists. It is the backbone of Pandas, Scikit-Learn, and TensorFlow.

```python
import numpy as np
import time

# Quick speed comparison
py_list = list(range(1000000))
np_arr = np.arange(1000000)

start = time.time()
[x * 2 for x in py_list]
print(f"Python List Time: {time.time() - start:.4f} seconds")

start = time.time()
np_arr * 2
print(f"NumPy Array Time: {time.time() - start:.4f} seconds")
```

## 2. Creating Arrays
Generating data structures from scratch or converting existing lists.

```python
# 1D Array
arr_1d = np.array([10, 20, 30, 40])

# 2D Array (Matrix)
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])

# Built-in Generators
zeros = np.zeros((3, 3))         # 3x3 matrix of 0.0
ones = np.ones((2, 4))           # 2x4 matrix of 1.0
evens = np.arange(0, 20, 2)      # Start 0, stop 20, step 2
dist = np.linspace(0, 100, 5)    # 5 evenly spaced numbers from 0 to 100
random_matrix = np.random.rand(3, 3) # 3x3 matrix of random floats 0-1

print("Evens Array:", evens)
```

## 3. Indexing & Slicing
Extracting specific elements or subsets of an array.

```python
arr = np.array([10, 20, 30, 40, 50, 60])

print("Index 2:", arr[2])         # 30
print("Last element:", arr[-1])   # 60
print("Slice [1:4]:", arr[1:4])   # [20, 30, 40]
print("Slice [::2]:", arr[::2])   # Step by 2: [10, 30, 50]

# Conditional Indexing (Boolean Masking)
print("Greater than 30:", arr[arr > 30]) # [40, 50, 60]
```

## 4. Multidimensional Indexing and Axis
In a 2D matrix, `axis=0` operates vertically (down rows), and `axis=1` operates horizontally (across columns).

```python
matrix = np.array([
    [10, 20, 30], 
    [40, 50, 60], 
    [70, 80, 90]
])

print("Row 1, Col 2:", matrix[1, 2]) # 60
print("Get entire Row 0:", matrix[0, :]) # [10, 20, 30]
print("Get entire Col 1:", matrix[:, 1]) # [20, 50, 80]

# Axis Operations
print("Sum down columns (axis=0):", matrix.sum(axis=0)) # [120, 150, 180]
print("Sum across rows (axis=1):", matrix.sum(axis=1))  # [60, 150, 240]
```

## 5. Data Types in NumPy
Forcing data types optimizes memory, crucial for large datasets.

```python
# Default types
arr_float = np.array([1.5, 2.5, 3.5])
print("Original Type:", arr_float.dtype) # float64

# Convert (cast) to integer
arr_int = arr_float.astype(np.int32)
print("Cast to Int:", arr_int)           # [1, 2, 3]
```

## 6. Broadcasting
Broadcasting lets NumPy do math on arrays of different shapes by "stretching" the smaller one.

```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])
scalar = 10
row_vector = np.array([100, 200, 300])

print("Broadcast Scalar:\\n", matrix * scalar)
# [[10, 20, 30], [40, 50, 60]]

print("\\nBroadcast Vector:\\n", matrix + row_vector)
# [[101, 202, 303], [104, 205, 306]]
```

## 7. Mathematical Functions

```python
arr = np.array([1, 4, 9, 16, 25])

print("Square Root:", np.sqrt(arr))
print("Mean:", np.mean(arr))
print("Standard Dev:", np.std(arr))
print("Max value index (argmax):", np.argmax(arr)) # Returns index 4
```

Modifying existing columns or creating new ones based on logic.

```python
# Using .map() to standardize string cases
df_clean['experience_level'] = df_clean['experience_level'].str.upper()

# Using .apply() with a lambda function for custom logic
df_clean['is_influencer'] = df_clean['connections'].apply(lambda x: True if x >= 100 else False)

display(df_clean)
```

## 5. Melt & Pivot
Reshaping data. Pivot goes Long-to-Wide, Melt goes Wide-to-Long.

```python
# Pivot Table Example: Average connections per experience level
pivot_df = df_clean.pivot_table(
    values='connections', 
    index='experience_level', 
    aggfunc='mean'
)
display(pivot_df)

# Melt Example: Unpivoting
melted_df = pd.melt(
    df_clean, 
    id_vars=['user_id', 'name'], 
    value_vars=['connections', 'is_influencer'],
    var_name='Metric', 
    value_name='Value'
)
display(melted_df.head(4))
```

## 6. Aggregation & Grouping
Summarizing data categories.

```python
# Group by experience and calculate multiple aggregates
summary_stats = df_clean.groupby('experience_level').agg({
    'user_id': 'count',                 # Count users
    'connections': ['mean', 'max']      # Avg and Max connections
})

# Rename columns for clarity
summary_stats.columns = ['Total Users', 'Avg Connections', 'Max Connections']
display(summary_stats)
```
