# Part 12: Large Language Models (LLMs) & AI integration
Large Language Models have revolutionized how Data Scientists process text, write code, and build applications.

## 1. How LLMs Work
* **History:** Evolved from simple Recurrent Neural Networks (RNNs) to the breakthrough **Transformer Architecture** (introduced by Google in 2017 in the paper *"Attention Is All You Need"*).
* **Mechanism:** LLMs don't "think" like humans; they predict the next most mathematically probable token (word/subword) based on massive amounts of training data.


[Image of Large Language Model Transformer Architecture]


## 2. Retrieval-Augmented Generation (RAG)
LLMs have a knowledge cutoff and can hallucinate. **RAG** solves this by connecting the LLM to an external database (like your own PDFs or SQL databases). 
1. **Retrieve:** Search the database for relevant information.
2. **Augment:** Combine the user's prompt with the retrieved data.
3. **Generate:** The LLM reads the combined data and generates an accurate, factual answer.

# Part 13: Git & GitHub for Data Scientists
Version control is mandatory for professional Data Science teams. It prevents the dreaded `final_model_v3_REAL_FINAL.ipynb` file naming disaster.


## 1. Core Git Workflow
* **Working Directory:** Where your modified files live.
* **Staging Area (`git add`):** A pre-commit holding area.
* **Local Repo (`git commit`):** Your saved snapshot on your hard drive.
* **Remote Repo (`git push`):** The cloud backup (GitHub).

## 2. Branches, Merging, & Stash
* **Branches (`git branch feature-x`):** Allow you to safely experiment with a new ML model without breaking your main code.
* **Merge Conflicts:** Happen when two people edit the same line of code. VS Code has a built-in UI to help you click "Accept Current" or "Accept Incoming".
* **Git Stash (`git stash`):** Temporarily saves your messy, uncommitted code so you can switch branches quickly without losing your work.

## 3. .gitignore
A file that tells Git what *not* to track. Crucial for Data Science!

```python
%%bash
# Example of what a Data Scientist's .gitignore looks like:
echo "*.csv" > .gitignore
echo "*.db" >> .gitignore
echo ".env" >> .gitignore
echo "__pycache__/" >> .gitignore
echo "*.pkl" >> .gitignore

echo "✅ Created a standard .gitignore for Data Science"
```
